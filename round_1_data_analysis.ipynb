```python
import pandas as pd
import matplotlib.pyplot as plt

pd1 = pd.read_csv("Data/ROUND1/prices_round_1_day_-1.csv", sep = ';')
pd2 = pd.read_csv("Data/ROUND1/prices_round_1_day_-2.csv", sep = ';')
```

```python
root_prices = pd1.loc[pd1['product'] == 'INTARIAN_PEPPER_ROOT', ['timestamp', 'mid_price', 'bid_price_1', 'ask_price_1']]
plt.plot(root_prices['timestamp'], root_prices['ask_price_1'])
```

```python
root_prices = pd2.loc[pd2['product'] == 'INTARIAN_PEPPER_ROOT', ['timestamp', 'mid_price', 'bid_price_1', 'ask_price_1']]
plt.plot(root_prices['timestamp'] < 600, root_prices['ask_price_1'])
plt.plot(root_prices['timestamp'] < 600, root_prices['mid_price'], color = 'red')
plt.plot(root_prices['timestamp'] < 600, root_prices['bid_price_1'], color = 'green')
```
